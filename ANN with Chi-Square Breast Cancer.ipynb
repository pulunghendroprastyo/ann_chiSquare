```python
#--------------------------------------------
#Created by Pulung Hendro Prastyo
#22 Desember 2019
#--------------------------------------------

#mengimport library yang dibutuhkan
#mengimport numpy, pandas dan keras
import numpy as np
import pandas as pd

#Import lib keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils.np_utils import to_categorical
from keras.optimizers import SGD #Gradient Discent
from keras.optimizers import adam #Gradient Discent v2

#Dataset
from sklearn.datasets import load_breast_cancer

#Untuk split data dan evaluasi
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics.classification import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
```

```python
#Dataset
dataset = load_breast_cancer()
X, y = dataset.data, dataset.target
features = dataset.feature_names
inputTraining, inputTesting, labelTraining, labelTesting = train_test_split(X, y, test_size=0.2, random_state=0)

labelTraining = to_categorical(labelTraining,num_classes=2)

```

```python
from datetime import datetime
start_time = datetime.now()

#mendefinisikan klasifier NN
inputSize = inputTraining.shape[1]
print(inputSize)

model = Sequential()
model.add(Dense(units=5, input_dim=inputSize)) # node input sesuai inputan 30 atribut/fitur dan hidden layer 5 node
#model.add(Activation('sigmoid'))
#model.add(Dense(units=10))
model.add(Activation('sigmoid')) #fungsi aktivasi menggunakan sigmoid
model.add(Dense(units=2)) #output layer 2 node
model.add(Activation('sigmoid'))
model.compile(loss='categorical_crossentropy',
             optimizer=adam(lr=0.001),
             metrics=['accuracy'])
#train NN
mlp=model.fit(inputTraining, labelTraining,
         epochs=150, batch_size=10)



#Time
##---Program--#
end_time = datetime.now()
result_time  =end_time-start_time
print("Duration:",result_time)
```

```python
#memprediksi data testing dengan model NN yg sdh ditraining

hasilPrediksi = model.predict_classes(inputTesting,batch_size=10)

#menghitung akurasi
print("\n-------------------------------------------")
print("label sebenarnya ", labelTesting)
print("hasil prediksi ", hasilPrediksi)
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("prediksi benar: ", prediksiBenar, " data")
print("prediksi salah: ", prediksiSalah, " data")
print("akurasi: ", prediksiBenar/(prediksiBenar+prediksiSalah)
     * 100, "%")
```

```python
CM = confusion_matrix(hasilPrediksi,labelTesting)
TN = CM[0][0]
FN = CM[1][0]
TP = CM[1][1]
FP = CM[0][1]
sensitivitas = TP/(TP+FN)
spesifisitas = TN/(FP+TN)
precision    = TP/(TP+FP)
recall       = TP/(TP+FN)
f1_score     = (2*(precision*recall))/(precision+recall)
auc          = roc_auc_score(hasilPrediksi,labelTesting)
print ("TRUE NEGATIVE (TN):",TN)
print ("FALSE NEGATIVE (FN):",FN)
print ("TRUE POSITIVE (TP):",TP)
print ("FALSE NEGATIVE (FP):",FP)
print ("SENSITIVITAS:",sensitivitas*100,"%")
print ("SPESIFISITAS:",spesifisitas*100,"%")
print ("PRECISION:",precision*100,"%")
print ("RECALL:",recall*100,"%")
print ("AUC:",auc*100,"%")
print ("F1-Score:",f1_score*100,"%")
print(classification_report(hasilPrediksi,labelTesting))
```
